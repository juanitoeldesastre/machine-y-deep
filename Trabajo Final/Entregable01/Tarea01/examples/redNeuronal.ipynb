```python
import tensorflow as tf
import numpy as np
```

# Creamos los datos de entrenamiento: grados Celsius y sus equivalentes en Fahrenheit

```python
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

# Definimos la red neuronal:
# - Una capa densa con una unidad de salida
# - Input shape de 1 porque la entrada es un solo número (un valor de Celsius)

```python
capa = tf.keras.layers.Dense(units=1, input_shape=[1])
# Creamos el modelo con la capa definida
modelo = tf.keras.Sequential([capa])
```

# Compilamos el modelo:
# - Usamos el optimizador 'Adam' con tasa de aprendizaje de 0.1
# - La función de pérdida es el error cuadrático medio (MSE), adecuado para regresión

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

# Entrenamos el modelo

```python
print("Comenzando entrenamiento...")
historial = modelo.fit(celsius, fahrenheit, epochs=1000, verbose=False)
print("Modelo entrenado!")
```

# Visualizamos la magnitud de la pérdida durante el entrenamiento

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(historial.history["loss"])
```

# Realizamos una predicción: convertir 100 grados Celsius a Fahrenheit

```python
# Realizamos una predicción: convertir 100 grados Celsius a Fahrenheit
print("Hagamos una predicción")
resultado = modelo.predict(np.array([100.0]))
print("El resultado es " + str(resultado) + " fahrenheit!")
```
